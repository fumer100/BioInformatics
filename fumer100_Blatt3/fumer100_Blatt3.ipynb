```python
#A1 [LGT mit Backtracking]
import numpy as np
def lgt(v,w):
    S = np.zeros(((len(v)+1),(len(w)+1)))
    B = np.empty(((len(v)+1),(len(w)+1)),dtype=object)
    for a in range(0,1):
        for b in range(1,len(w)+1):
            B[a,b]="links"
            
    for c in range(1,len(v)+1):
        for d in range(0,1):
            B[c,d]="oben"
            
    for i in range(1,len(v)+1):
        for j in range(1,len(w)+1):
            if(v[i-1]==w[j-1]):
                S[i,j]=max(S[i-1,j],S[i,j-1],S[i-1,j-1]+1)
                if(max(S[i-1,j],S[i,j-1],S[i-1,j-1]+1)==S[i-1,j]):
                    B[i,j]="oben"
                if(max(S[i-1,j],S[i,j-1],S[i-1,j-1]+1)==S[i,j-1]):
                    B[i,j]="links"
                if(max(S[i-1,j],S[i,j-1],S[i-1,j-1]+1)==S[i-1,j-1]+1):
                    B[i,j]="oben-links"
            else:
                S[i,j]=max(S[i-1,j],S[i,j-1])
                if(S[i-1,j]==S[i,j-1]):
                    B[i,j]="oben/links"
                elif(max(S[i-1,j],S[i,j-1])==S[i-1,j]):
                    B[i,j]="oben"
                elif(max(S[i-1,j],S[i,j-1])==S[i,j-1]):
                    B[i,j]="links"
    
    score=S[(len(v))][(len(w))]
    print(S)
    print(B)
    return (score,B)

def backtrack_v(B,v,i,j):
    v_lgt=list(v)
    if i==0 and j==0:         
        print(v_lgt)
        return(v_lgt)
    
    if B[i,j]=="oben/links":
        v_lgt[i-1]="-"
        backtrack_v(B,v_lgt,i,j-1)
    if B[i,j]=="links":
        v_lgt[i-1]="-"
        backtrack_v(B,v_lgt,i,j-1)
    if B[i,j]=="oben":
        backtrack_v(B,v_lgt,i-1,j)
    if B[i,j]=="oben-links":
        backtrack_v(B,v_lgt,i-1,j-1)
        
def backtrack_w(B,w,i,j):
    w_lgt=list(w)
    if i==0 and j==0:         
        print(w_lgt)
        return(w_lgt)
    
    if B[i,j]=="oben/links":
        w_lgt[j-1]="-"
        backtrack_w(B,w_lgt,i-1,j)
    if B[i,j]=="links":
        backtrack_w(B,w_lgt,i,j-1)
    if B[i,j]=="oben":
        w_lgt[j-1]="-"
        backtrack_w(B,w_lgt,i-1,j)
    if B[i,j]=="oben-links":
        backtrack_w(B,w_lgt,i-1,j-1)
        
v = "ATGCG"
w = "ATCCGC"
S, B = lgt(v,w)
backtrack_v(B,v,len(v),len(w))
backtrack_w(B,w,len(v),len(w))
```
