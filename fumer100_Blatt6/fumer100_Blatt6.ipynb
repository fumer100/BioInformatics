```python
#A3 PruneAndRegraft
def pruneAndRegraft(tree,prune,regraft):
    pruned, toBeInserted = prune_at(tree,prune)
    print('pruned at :',prune)
    print(list(pruned))
    regrafted= regraft_at(pruned, toBeInserted,regraft,prune)
    print('regrafted at :', regraft)
    print(regrafted)
    return regraft_at(pruned, toBeInserted,regraft,prune)


def prune_at(tree,prune):
    pruned_tree=tree.copy()
    pruned_tree.remove(prune)
    badTwigs=[]
    toBeDeleted=[]
    leftleaf=prune[0]
    rightleaf=prune[1]
    for twig in pruned_tree: 
        if rightleaf==twig[0]:
            badTwigs.append(twig)
            
            
    for bad in badTwigs:
        toBeDeleted.append(bad)
        for twig in pruned_tree:
            if bad[1]==twig[0]:
                toBeDeleted.append(twig)
             
    for twig in pruned_tree: #the prune containing tuples
        
        if leftleaf==twig[0]:
            pruned_tree.remove(twig)
            for twig1 in pruned_tree:
                if twig1[1]==twig[0]:
                    pruned_tree.append((twig1[0],twig[1]))
        if leftleaf==twig[1]:
            pruned_tree.remove(twig)
    
         
    pruned=set(pruned_tree)-set(toBeDeleted)    
    return (pruned,set(toBeDeleted)) 



def regraft_at(prunedTree,toBeInserted,regraft,prune):
    regraftedTree=list(prunedTree).copy()
    regraftedTree.remove(regraft)
    regraftedTree.append((regraft[0],prune[0]))
    regraftedTree.append((prune[0],regraft[1]))
    
    for twig in toBeInserted:
        regraftedTree.append(twig)

    return regraftedTree



tree=[(10, 1), (10, 2), (11, 4), (11, 5), (12, 6), (12, 7), (13, 11), (13, 12), (14, 3), (14, 13), (15, 8), (15, 9), (16, 14), (16, 15), (17, 10), (17, 16)]
pruneAndRegraft(tree,(14,13),(15,8))
```
